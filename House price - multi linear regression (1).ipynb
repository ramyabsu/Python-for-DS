```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('House Price.csv')
```

```python
dataset
```

```python
print(len(dataset))
print(dataset.isnull().any())
dataset.isnull().sum()
```

```python
#check for missing value, unique etc
new_dataframe = pd.DataFrame(columns=['column_name','missing_value_count','percent_missing_value','unique_value_count'])
for col in list(dataset.columns.values):
    sum_missing = dataset[col].isnull().sum()
    percent_missing = sum_missing/len(dataset)*100
    uniq_count = (dataset.groupby([col])[col].count()).count()
    new_dataframe = new_dataframe.append({'column_name':col,'missing_value_count':sum_missing,
                                        'percent_missing_value':percent_missing,'unique_value_count':uniq_count},
                                       ignore_index = True)
```

```python
new_dataframe
```

```python
File_Col_Type = (dataset.dtypes).reset_index()
```

```python
#merging Column Type with other information about dataset
new_dataframe = new_dataframe.merge(File_Col_Type, left_on = 'column_name', right_on = 'index', how = 'inner')
```

```python
del new_dataframe['index']
new_dataframe = new_dataframe.rename(index=str, columns = {0:"column_type"})
```

```python
new_dataframe
```

```python
#Drop column with more than 25% null values
dataset.drop(['Alley'], axis = 1, inplace = True)
```

```python
#Apply Mode strategy to populate the categorical data
dataset.groupby('GarageType').agg({'GarageType': np.size})
dataset["GarageType"]=dataset["GarageType"].fillna('Attchd')
#Apply encoder
dataset["GarageType"] = dataset["GarageType"].astype('category')
dataset["GarageType"] = dataset["GarageType"].cat.codes
```

```python
#Mean strategy to populate the quantitative variable
dataset["LotFrontage"].fillna(dataset["LotFrontage"].mean(), inplace=True)
```

```python
#Median strategy to populate the quantitative variable
dataset["MasVnrArea"].fillna(dataset["MasVnrArea"].median(), inplace=True)
```

```python
dataset = pd.get_dummies(dataset, columns=['GarageType'], drop_first=True)
```

```python
dataset_salesPrice=dataset['SalePrice']
```

```python
dataset.drop(['SalePrice'], axis = 1, inplace = True)
```

```python
dataset=pd.concat([dataset,dataset_salesPrice],axis=1)
dataset
```

```python
dataset.isnull().sum()
```

```python
#Analyse the selected variables
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(dataset)
```

```python
x = dataset.iloc[:,1:13]
y = dataset.iloc[:,13:]
```

```python
# split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
my_regressor = LinearRegression()
my_regressor.fit(x_train, y_train)
```

```python
my_regressor.coef_
```

```python
my_regressor.intercept_
```

```python
y_pred = my_regressor.predict(x_test)
```
