```python
#missing value treatment and encoding 

#from sklearn.impute import SimpleImputer
#imp = SimpleImputer(strategy='most_frequent')
#or

# Gender
dataset_xtrain["Gender"]=dataset_xtrain["Gender"].fillna('Male')
dataset_xtrain["Gender"] = dataset_xtrain["Gender"].astype('category')
dataset_xtrain["Gender"] = dataset_xtrain["Gender"].cat.codes


# Married
dataset_xtrain["Married"]=dataset_xtrain["Married"].fillna('Yes')
dataset_xtrain["Married"] = dataset_xtrain["Married"].astype('category')
dataset_xtrain["Married"] = dataset_xtrain["Married"].cat.codes
#Dependents
dataset_xtrain["Dependents"]=dataset_xtrain["Dependents"].fillna('0')
dataset_xtrain.replace('3+', 3, inplace=True)
# Education ( no missing value but only encoding)
dataset_xtrain["Education"] = dataset_xtrain["Education"].astype('category')
dataset_xtrain["Education"] = dataset_xtrain["Education"].cat.codes
# Self_Employed 
dataset_xtrain["Self_Employed"]=dataset_xtrain["Self_Employed"].fillna('Yes')
dataset_xtrain["Self_Employed"] = dataset_xtrain["Self_Employed"].astype('category')
dataset_xtrain["Self_Employed"] = dataset_xtrain["Self_Employed"].cat.codes
# LoanAmount
dataset_xtrain["LoanAmount"].fillna(dataset_xtrain["LoanAmount"].mean(), inplace=True)
# Loan_Amount_Term
dataset_xtrain["Loan_Amount_Term"]=dataset_xtrain["Loan_Amount_Term"].fillna('360.0')
#Credit_History
dataset_xtrain["Credit_History"]=dataset_xtrain["Credit_History"].fillna('1')
# Property_Area
dataset_xtrain["Property_Area"] = dataset_xtrain["Property_Area"].astype('category')
dataset_xtrain["Property_Area"] = dataset_xtrain["Property_Area"].cat.codes
# Loan_Status
dataset_xtrain["Loan_Status"] = dataset_xtrain["Loan_Status"].astype('category')
dataset_xtrain["Loan_Status"] = dataset_xtrain["Loan_Status"].cat.codes
```
